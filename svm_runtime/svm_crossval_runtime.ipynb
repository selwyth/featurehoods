```python
import pandas as pd
from seaborn import plt
%matplotlib inline

df = pd.read_csv('svm_crossval_runtime.csv')
df.head()
```

```python
import numpy as np

df['logC'] = df.C.apply(np.log10)
df.head()
```

```python
import seaborn as sns
sns.lmplot('logC', 'runtime', df, hue='gamma', legend=True)
plt.ylim(0,500)
plt.xlim(2.5, 7)
# sns.regplot('C', 'runtime', df, scatter=True, color = 'gamma', logx=True)
```
